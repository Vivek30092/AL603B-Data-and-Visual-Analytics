```python
import pandas as pd 
import numpy as np 

# read file
df = pd.read_csv('Superstore_USA.csv',encoding='cp1252')
df.head()
```

```python
# convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

#remove duplicate rows
df = df.drop_duplicates()      #our data does't have missing values

# drop missing values
df = df.dropna()
```

### 3. Find out how many different shipment modes that we have and what is a percentage usability of all the shipment mode with respect to dataset

```python
shipment_counts = df['Ship Mode'].value_counts()
print(shipment_counts)
```

```python
# calculate percentage usage
shipment_percentage = (shipment_counts / len(df)) * 100

result = pd.DataFrame({
    'Count': shipment_counts,
    'Percentage (%)': shipment_percentage
})

print(result)
```

### 4. Create a new column and try to find out a difference between order date and shipment date

```python
df['Shipment Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

# Preview
df[['Order ID', 'Order Date', 'Ship Date', 'Shipment Duration']].head()
```

### 5. Base on question number 8 find out for which order id we have shipment duration more than 10 days

### 7. Gorup by region and find out which region is more profitable

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

print(region_profit)
```

### 8. Try to find out overall in which country we are giving more discount

```python
country_discount = df.groupby('State or Province')['Discount'].sum().sort_values(ascending=False)
```

```python
top_state = country_discount.idxmax()
print("State with highest discount:", top_state)
```

### 10. Which customer segment is more profitable find it out.


```python
segment_profit = df.groupby('Customer Segment')['Profit'].sum().sort_values(ascending=False)

print(segment_profit)
```

```python
top_segment = segment_profit.idxmax()
print("Most Profitable Segment:", top_segment)
```

### 11. Try to find out the 10th most loss-making product category.

```python
category_profit = df.groupby('Product Category')['Profit'].sum().sort_values()
print(category_profit)
```

```python
if len(category_profit) >= 10:
    tenth_loss = category_profit.iloc[9]
    print("10th most loss-making category:", tenth_loss)
else:
    print("Not enough categories to find 10th loss-making")
```

### 12. Try to find out 10 top products with highest margins.

```python
df['Margin'] = df['Profit'] / df['Sales']
```

```python
top_products = df.groupby('Product Name')['Margin'].mean().sort_values(ascending=False).head(10)

print(top_products)
```
